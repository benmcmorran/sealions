```python
import glob
import numpy as np
import cv2
import matplotlib.pyplot as plt
import json
%matplotlib notebook
```

```python
with open('sliced_small_instances.json') as in_f:
    instances = json.load(in_f)
    images = instances['images']
    annotations = instances['annotations']
```

```python
ann = annotations[204]
img_id = ann['image_id']
img = cv2.imread('./data/Sliced/{0}'.format(img_id))
img = cv2.cvtColor(img, cv2.cv.CV_BGR2RGB)
anns = [a for a in annotations if a['image_id'] == img_id]
for sub_ann in anns:
    bb = sub_ann['bbox']
    cv2.rectangle(img,(bb[1],bb[0]),(bb[1]+50,bb[0] + 50),(255,0,0),2)
plt.close()
plt.imshow(img)
cv2.imwrite('test.jpg', img)
```
